```python
import math
import numpy as np
import IPython.display
```

```python
# Taked from music from geeks https://pedrokroger.net/mfgan/ Chapter 6.1 The basics of sound
semitone_constant = 2**(1/12)
semitone_constant
```

```python
notes = ["A", "A#", "B", "C", "C#", "D", "D#", "E", "F", "F#", "G", "G#"]
len(notes)
```

```python
def freq_2_note(x):
    interval = int(round(math.log(x/440, semitone_constant))) % 12
    return notes[interval]

def note_2_freq(n):
    if n in notes:
        return 440 * (semitone_constant ** notes.index(n))
```

```python
note_2_freq("C")
```

```python
sampling_frequency = 16000
def wave_generator(f):
    t = np.linspace(0,sampling_frequency,sampling_frequency)
    y = np.sin(2*np.pi*f*t/sampling_frequency)
    return IPython.display.Audio(y, rate=sampling_frequency)
```

```python
wave_generator(note_2_freq("A"))
```

```python
wave_generator(110)
```

```python
wave_generator(262)
```
