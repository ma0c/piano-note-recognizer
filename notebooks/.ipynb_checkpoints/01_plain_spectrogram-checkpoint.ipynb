```python
import numpy as np
from scipy.io import wavfile
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
import IPython.display
```

```python
# First we load a real audio
frequency, wave = wavfile.read("../wav/C.wav")
```

```python
plt.plot(wave)
```

```python
IPython.display.Audio(wave, rate=frequency)
```

```python
# And calculate the spectrgram
frequencies, time, spectrogram = signal.spectrogram(wave, frequency)
cmap=get_cmap("coolwarm")
plt.pcolormesh(time, frequencies, spectrogram,cmap=cmap)
```

```python
duration = wave.shape[0]/frequency
duration
```

```python
# Now, lets generate a virtual one
f = 16000
t = np.linspace(0,f*duration,int(f*duration))
y = np.sin(2*np.pi*262*t/f)
plt.plot(t,y)
```

```python
IPython.display.Audio(y, rate=f)
```

```python
# The corresponding spectrogram
frequencies, time, spectrogram = signal.spectrogram(y, f)
plt.pcolormesh(time, frequencies, spectrogram,cmap=cmap)
```

```python
f_per, Pxx = signal.periodogram(wave, frequency)
plt.plot(f_per,Pxx)
```

```python
# Lets calculate the power over the original signal
n_chunks = 300
chunks = np.array_split(wave.astype('int64'), n_chunks)
len(chunks)
```

```python
power = lambda x: np.sum(np.square(np.abs(x)))/x.shape[0]
```

```python
power(chunks[0])
```

```python
list_of_powers = np.apply_along_axis(power, 1, chunks, )
list_of_powers.shape
```

```python
samples_per_chunk = chunks[0].shape
samples_per_chunk
```

```python
plt.plot(list_of_powers)
```

```python
normalized_power = list_of_powers / np.linalg.norm(list_of_powers)
plt.plot(normalized_power)
```

```python
np.max(normalized_power)
```

```python
normalized_power.shape
```

```python
interpolated_power = np.interp(t,np.linspace(0,t.shape[0],100),normalized_power)
interpolated_power.shape
```

```python
plt.plot(interpolated_power)
```

```python
generated_wave_with_same_shape = y*interpolated_power
```

```python
plt.plot(generated_wave_with_same_shape)
```

```python
generated_wave_with_same_shape.shape
```
